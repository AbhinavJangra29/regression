```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class mylinreg:
  def __init__(self):
    self.w=None;
    self.b=None;

  #fit the line
  def fit(self,x,y):
    n=len(x)
    x_sum=np.sum(x)
    y_sum=np.sum(y)
    xy_sum=np.sum(x*y)
    xsquare_sum=np.sum(x**2)
    numerator = n * x_sum - xy_sum * y_sum
    denominator = n * xsquare_sum - x_sum**2

    self.w=numerator/denominator
    self.b=(y_sum-self.w*x_sum)/n

  #predict
  def predict(self,x):
    return self.w*x+self.b
```

```python
X = np.array([1, 2, 3, 4, 5])
Y = np.array([3, 6, 4, 7, 9])
```

```python
model=mylinreg()
```

```python
model.fit(X,Y)
```

```python
pred_y=model.predict(X)
```

```python
plt.scatter(X, Y, label="Original Data")
plt.plot(X, pred_y, color='red', label="Fitted Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
