```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class optlr:
  def __init__(self,alpha=0.01,itr=1000):
    self.alpha=alpha
    self.itr=itr
    self.w=None
    self.b=None

  def fit(self,x,y):
    n=len(x);
    self.w=np.random.randn()
    self.b=np.random.randn()

    for _ in range(self.itr):
      y_pred=self.w*x+self.b
      error=y_pred-y
      grad_w=(2/n)*np.dot(error,x)
      grad_b=(2/n)*np.sum(error)
      self.w -=self.alpha*grad_w
      self.b -=self.alpha*grad_b

   #predict
  def predict(self,x):
    return self.w*x+self.b
```

```python
X = np.array([1, 2, 3, 4, 5])
Y = np.array([3, 6, 4, 7, 9])
```

```python
newmodel=optlr(alpha=0.01 ,itr=10000)
```

```python
newmodel.fit(X,Y)
```

```python
start_x = min(X)
end_x = max(X)
start_y = newmodel.predict(start_x)
end_y = newmodel.predict(end_x)
```

```python
plt.scatter(X, Y, label="Original Data")
plt.plot([start_x, end_x], [start_y, end_y], color='red', label="Best-Fit Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression - Best-Fit Line (Gradient Descent)")
plt.legend()
plt.show()
```
